```python
# Trim any leading or trailing spaces in column names
df.columns = df.columns.str.strip()

# Convert 'distancia_viaje_km' to numeric (handle potential string values)
df["distancia_viaje_km"] = pd.to_numeric(df["distancia_viaje_km"], errors="coerce")

# Filter dataset for home-based work trips within Buenos Aires City (CABA)
filtered_df = df[
    (df["trip_type"] == "CABA") &  # Trips within Buenos Aires City
    (df["trip_reason"] == 3)  # Assuming '3' corresponds to work trips
]

# Calculate the proportion of trips where the distance is under 1 km
proportion_under_1km = (filtered_df["distancia_viaje_km"] < 1).mean()

# Display result
print(f"Proportion of trips under 1km: {proportion_under_1km:.2%}")
```
