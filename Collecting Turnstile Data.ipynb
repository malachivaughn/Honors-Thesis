```python
import pandas as pd
import os
import glob
import io

# Define folder path
folder_path = "BA Turnstile Data"
output_file = "BA_Turnstile_Combined.csv"

# Get all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# List to store DataFrames
dfs = []
index = 1

# Process each CSV file
for file in csv_files:
    print(index)
    with open(file, 'r', encoding='utf-8-sig') as f:
        lines = [line.strip().rstrip(';') for line in f]  # Remove trailing semicolons

    # Convert cleaned content into a readable format
    cleaned_file = io.StringIO("\n".join(lines))

    # Read the CSV
    df_raw = pd.read_csv(cleaned_file, delimiter=";", encoding="utf-8-sig", header=None)

    # Extract column names from the first row
    column_names = df_raw.iloc[0, 0].split(";")  
    df = df_raw.iloc[1:].copy()  # Skip the header row

    # Split first column into multiple columns
    df.iloc[:, 0] = df.iloc[:, 0].str.split(";")
    df = df.iloc[:, 0].apply(pd.Series)

    # Assign correct column names
    df.columns = column_names
    df = df.reset_index(drop=True)

    # Store the processed DataFrame
    dfs.append(df)
    index += 1

# Combine all DataFrames
final_df = pd.concat(dfs, ignore_index=True)

# Save the final combined CSV
final_df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(f"✅ Combined CSV saved as: {output_file}")
```
