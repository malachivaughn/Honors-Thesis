```python
import numpy as np

# Example: 4 stations with total trips recorded
stations = ["A", "B", "C", "D"]
total_trips = np.array([2150, 3050, 2200, 2800])  # Hypothetical total trips per station

# Generate variable mappings for bidirectional flows
station_pairs = [(i, j) for i in range(len(stations)) for j in range(i + 1, len(stations))]
num_vars = len(station_pairs)

# Construct matrix A (each row represents a station, each column a bidirectional trip)
A = np.zeros((len(stations), num_vars))

for station_idx, station in enumerate(stations):
    for var_idx, (i, j) in enumerate(station_pairs):
        if station_idx in (i, j):
            A[station_idx, var_idx] = 1  # This trip contributes to the station's total count

# Solve using least squares
X, residuals, rank, s = np.linalg.lstsq(A, total_trips, rcond=None)

# Compute error: how wrong is the estimation?
estimated_trips = A @ X
error = total_trips - estimated_trips  # Difference between actual and estimated totals
error_norm = np.linalg.norm(error)

# Display results
print("Station Pair Flow Estimates:")
for (i, j), flow in zip(station_pairs, X):
    print(f"{stations[i]} <-> {stations[j]}: {flow:.2f}")

print("\nResidual Errors (Actual - Estimated):")
for station, err in zip(stations, error):
    print(f"{station}: {err:.2f}")

print(f"\nTotal Residual Error Norm: {error_norm:.2f}")
```

```python
import numpy as np
from scipy.optimize import minimize

# Generate unique station pairs for bidirectional flows
num_stations = len(station_list)
station_pairs = [(i, j) for i in range(num_stations) for j in range(i + 1, num_stations)]
num_vars = len(station_pairs)

# Construct matrix A dynamically
A = np.zeros((num_stations, num_vars))

for station_idx in range(num_stations):
    for var_idx, (i, j) in enumerate(station_pairs):
        if station_idx in (i, j):
            A[station_idx, var_idx] = 1  # This trip contributes to the station's total count

# Define objective function for quadratic optimization
def objective(X):
    return np.linalg.norm(A @ X - b_array) ** 2  # Minimize squared error

# Solve using constrained optimization (non-negative values)
X0 = np.ones(num_vars)  # Initial guess (small nonzero values to prevent collapse)
bounds = [(0, None) for _ in range(num_vars)]  # Non-negative constraints

res = minimize(objective, X0, bounds=bounds, method='L-BFGS-B')

X_opt = res.x  # Optimal trip flow values

# Compute final error
final_estimated_trips = A @ X_opt
final_error = b_array - final_estimated_trips
final_error_norm = np.linalg.norm(final_error)

# Display results
print("Station Pair Flow Estimates:")
for (i, j), flow in zip(station_pairs, X_opt):
    print(f"{station_list[i]} <-> {station_list[j]}: {flow:.2f}")

print("\nResidual Errors (Actual - Estimated):")
for station, err in zip(station_list, final_error):
    print(f"{station}: {err:.2f}")

print(f"\nTotal Residual Error Norm: {final_error_norm:.2f}")
```

```python
commuter_array = np.zeros([78,78])

for i in range(3003):
    joined_total = solution[i]
    
    station1_index = station_pairs[i][0]
    station2_index = station_pairs[i][1]
    
    station1 = station_list[station1_index]
    station2 = station_list[station2_index]
    
    station1_pop = pop_dict[station1]
    station2_pop = pop_dict[station2]
    total_pop = station1_pop + station2_pop
    
    commuter_array[station1_index][station2_index] = joined_total * station1_pop/total_pop
    commuter_array[station2_index][station1_index] = joined_total * station2_pop/total_pop
    
commuter_array
```
